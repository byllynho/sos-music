```python
import csv
from nltk.corpus import stopwords
from nltk import word_tokenize

songs_dictionary = {}
l_lyric = ''
all_topics = []

with open('csv_songs_topics.csv', mode='r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    irrelevant = ['Hope','Selfishness','Psychotic','Narcissist','Prostitution','Personal-Style','Capitalism','Masculism','Maniac','Dissapointment', 'Chaos', 'Riot']
    line_count = 0
    for row in csv_reader:
        if row['description'] not in irrelevant:
            if row['description'] not in all_topics:
                all_topics.append(row['description'])

            if row['id'] in songs_dictionary:
                songs_dictionary[row['id']]['topics'].append(row['description'])
            else:
                lyric = row['lyric'].encode('utf-8')
                tokens = word_tokenize(lyric)
                # convert to lower case
                tokens = [w.lower() for w in tokens]
                # remove remaining tokens that are not alphabetic
                words = [word for word in tokens if word.isalpha()]
                # filter out stop words
                from nltk.corpus import stopwords
                stop_words = set(stopwords.words('english'))
                keywords = [w for w in words if not w in stop_words]

                words = ' '.join(keywords)


                songs_dictionary[row['id']] = {'song': row['title'], 'lyrics': words, 'topics': [row['description']]}

            line_count += 1

            
print(songs_dictionary)


#print((songs_dictionary))   
```
